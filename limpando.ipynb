```python
import  pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
###criando df para trabalho

df_movies = pd.read_csv('ml-latest-small/movies.csv')

df_tags = pd.read_csv("ml-latest-small/tags.csv")


movies_tags = df_movies.merge(df_tags, on='movieId')



df_tags = df_tags.drop(columns=['userId', 'timestamp'])


df_tags_normalize = df_tags.groupby('movieId')['tag'].agg('|'.join).reset_index()


movies_tags = df_movies.merge(df_tags_normalize, on='movieId', how='left')


# movies_tags.to_csv('data/cbf_df.csv', index=False)
```

```python
df_movies_gen = df_movies['genres']


df_movies_gen.to_csv('generos.csv')
```

```python

colunas_para_combinar = [col for col in movies_tags.columns if col != 'movieId']

print(type(movies_tags))
movies_tags["geral"] = movies_tags[colunas_para_combinar].apply(
    lambda row: ' '.join(row.values.astype(str)), 
    axis=1
)

movies_tags["geral"] = movies_tags["geral"].apply(lambda x: x.replace(",", " "))
movies_tags["geral"] = movies_tags["geral"].apply(lambda x: x.replace("|", " "))
movies_tags["geral"] = movies_tags["geral"].apply(lambda x: x.replace("(", " "))
movies_tags["geral"] = movies_tags["geral"].apply(lambda x: x.replace(")", " "))
movies_tags.to_csv('data/cbf_df.csv', index=False)
```

```python
lista_compilado = movies_tags["geral"].tolist()

print(lista_compilado)
```

```python
cvec = CountVectorizer(stop_words='english', min_df=3, max_df=0.5, ngram_range=(1,2))

sf = cvec.fit_transform(lista_compilado)


print(sf)
```

```python
cvec = CountVectorizer(stop_words='english', min_df=3, max_df=0.5, ngram_range=(1,2))

sf = cvec.fit_transform(lista_compilado)


print(sf)
```

```python
transformer = TfidfTransformer()
tfidf_matrix = transformer.fit_transform(sf)
print(tfidf_matrix)
```

```python
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
def recomendar_filmes(titulo_filme, movies_df, cosine_sim_matrix, n_recomendacoes=10):
    print(f'gerando recomendações para filme {titulo_filme}\n')
    indices = pd.Series(movies_df.index, index=movies_df['titulo_movie_lens']).drop_duplicates()
    
    if titulo_filme not in indices:
        print(f"Filme '{titulo_filme}' não encontrado no dataset.")
        return pd.Series([])

    idx = indices[titulo_filme]

    sim_scores = list(enumerate(cosine_sim_matrix[idx]))



    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:n_recomendacoes + 1]  

    filme_indices = [i[0] for i in sim_scores]
    print(f"filmes similares ao {titulo_filme}")
    print(movies_df['titulo_movie_lens'].iloc[filme_indices].to_string(index=False))
    return movies_df['titulo_movie_lens'].iloc[filme_indices]



filme = "Toy Story (1995)"
recomendados = recomendar_filmes(filme, movies_tags, cosine_sim)
```
