```python
import pandas as pd
```

```python
df = pd.read_csv('teste2.csv')

df.pop('poster_url')
df.pop('movieId')
df.pop('title')
df.pop('year')

df
```

```python
df.loc[:, 'genres']

df_genero = pd.DataFrame(columns=['genero'])


def alinhando(string, df_ajustado):
    dividindo = string.split('|')
    df_novos = pd.DataFrame({'genero': dividindo})

    # 2. Concatenamos os dois
    df_ajustado = pd.concat([df_ajustado, df_novos], ignore_index=True)

    return df_ajustado


        
df_genero = alinhando('Adventure|Animation|Children|Comedy|Fantasy', df_genero)

df_genero

```

```python
def extrair_generos(string):
    if pd.isna(string):
        return []
    return string.split('|')

# 2. Aplicando a função na coluna 'genres'
# Isso criará uma coluna onde cada linha contém uma lista: ['Adventure', 'Animation', ...]
df['generos_lista'] = df['genres'].apply(extrair_generos)

# 3. O "Pulo do Gato": Usar o explode para transformar cada item da lista em uma nova linha
df_genero = df[['generos_lista']].explode('generos_lista')

# 4. Ajustar nomes e remover duplicatas (opcional)
df_genero.columns = ['genero']
df_genero = df_genero.reset_index(drop=True)

df_genero
```
