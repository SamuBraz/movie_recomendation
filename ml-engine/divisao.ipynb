```python
import  pandas as pd
import math
```

```python
df_ratings = pd.read_csv('ml-latest-small/ratings.csv')

### Pega apenas notas superiores a
df_ratings_hg = df_ratings.loc[df_ratings['rating'] >= 4]

### Ordena o timestamp :)
df_ratings_hg = df_ratings_hg.sort_values(by=['userId', 'timestamp'])
df_test = df_ratings_hg.groupby('userId').apply(lambda x: x.tail(math.ceil(len(x) * 0.1)))
```

```python
df_test.to_csv('data_spliting/testing.csv')
```

```python
df_train = df_ratings_hg.drop(df_test.index.get_level_values(-1))
```

```python
df_train.to_csv('data_spliting/traing.csv', index=False)
```

```python
df_ratings_hg.to_csv('data_spliting/total.csv', index=False)
```
