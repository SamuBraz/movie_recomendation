```python
import  pandas as pd
import math
```

```python
df_ratings = pd.read_csv('ml-latest-small/ratings.csv')

### Pega apenas notas superiores a
df_ratings_hg = df_ratings.loc[df_ratings['rating'] >= 3.5]

### Ordena o timestamp :)
df_ratings_hg = df_ratings_hg.sort_values(by=['userId', 'timestamp'])
df_test = df_ratings_hg.groupby('userId').apply(lambda x: x.tail(math.ceil(len(x) * 0.1)))
```

```python
df_test.to_csv('data_spliting3/testing.csv')
```

```python
df_train = df_ratings_hg.drop(df_test.index.get_level_values(-1))
```

```python
df_train.to_csv('data_spliting3/traing.csv', index=False)
```

```python
import  pandas as pd
import math
```

```python
###separando de forma regular e 80-20

df_ratings = pd.read_csv('ml-latest-small/ratings.csv')
df_ratings_hg = df_ratings.loc[df_ratings['rating'] >= 4.0]


# 2. Criamos um contador (rank) para cada avaliação de cada usuário
# Isso numera as linhas de 0 a N para cada userId
df_ratings_hg['rank'] = df_ratings_hg.groupby('userId').cumcount()


# 3. Seleção Escalonada (20% -> 1 a cada 5)
# Usamos o operador módulo (%). Se o resto da divisão por 5 for 0, vai para o teste.
mask_test = (df_ratings_hg['rank'] % 5 == 0)

df_test = df_ratings_hg[mask_test].drop(columns=['rank'])
df_train = df_ratings_hg[~mask_test].drop(columns=['rank'])

df_test.to_csv('data_spliting4/testing.csv', index= False)
df_train.to_csv('data_spliting4/traing.csv', index= False)
```
