```python
import pandas as pd
recomendados = pd.read_csv('recomenados-93.csv')
recomendados2 = pd.read_csv('recomenados-93-2.csv')

df_movies = pd.read_csv('data/cbf_df.csv')
df_movies = df_movies.drop(columns=['geral', 'tag'])


df_testing = pd.read_csv('data_spliting4/testing.csv')
df_testing = df_testing.loc[df_testing['userId'] == 93]
df_testing = df_testing.drop(columns=['rating','timestamp'])

df_traning = pd.read_csv('data_spliting4/traing.csv')
df_traning = df_traning.loc[df_traning['userId'] == 93]
df_traning = df_traning.drop(columns=['rating','timestamp'])




recomendados = recomendados.merge(df_movies, 'left', on='movieId').drop(columns=['titulo_movie_lens_y'])
recomendados = recomendados.rename(columns={'titulo_movie_lens_x': 'titulo_movie_lens'})

recomendados2 = recomendados2.merge(df_movies, 'left', on='movieId')
recomendados2 = recomendados2.rename(columns={'titulo_movie_lens_x': 'titulo_movie_lens'})

df_testing = df_testing.merge(df_movies, 'left', on='movieId')
df_traning = df_traning.merge(df_movies, 'left', on='movieId')
```

```python
dummies = recomendados['genres'].str.get_dummies(sep='|')
recomendados = pd.concat([recomendados, dummies], axis=1)
recomendados

dummies = recomendados2['genres'].str.get_dummies(sep='|')
recomendados2 = pd.concat([recomendados2, dummies], axis=1)
recomendados2


dummies = df_testing['genres'].str.get_dummies(sep='|')
df_testing = pd.concat([df_testing, dummies], axis=1)
df_testing

dummies = df_traning['genres'].str.get_dummies(sep='|')
df_traning = pd.concat([df_traning, dummies], axis=1)
df_traning
```

```python
recomendados_generos = recomendados.iloc[:, 3:]
recomendados_generos

df_traning_generos = df_traning.iloc[:, 4:]
df_traning_generos

recomendados_generos2 = recomendados2.iloc[:, 3:]
recomendados_generos2


recomendados_generos.sum().sort_values().to_csv('validation/soma_recomendados.csv')
df_traning_generos.sum().sort_values().to_csv('validation/soma_treino.csv')

recomendados_generos2.sum().sort_values().to_csv('validation/soma_recomendados2.csv')
```
