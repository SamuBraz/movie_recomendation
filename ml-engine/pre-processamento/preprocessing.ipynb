# Pré-processamento da coluna `geral`

Limpeza do texto antes de ser consumido pelo TF-IDF:
- Lowercase
- Remoção de pontuação
- Remoção de acentos

```python
import pandas as pd
import unicodedata
import re

df = pd.read_csv('data/cbf_df.csv')
print(df.shape)
df['geral'].head(3)
```

```python
def preprocess(text):
    if pd.isna(text):
        return ''
    # Lowercase
    text = text.lower()
    # Remove acentos
    text = unicodedata.normalize('NFD', text)
    text = ''.join(c for c in text if unicodedata.category(c) != 'Mn')
    # Remove pontuação (mantém espaços e alfanuméricos)
    text = re.sub(r'[^a-z0-9\s]', ' ', text)
    # Normaliza espaços múltiplos
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df['geral'] = df['geral'].apply(preprocess)
df['geral'].head(3)
```

```python
# Verificação: nenhuma linha com NaN ou string vazia
print("Vazios:", df['geral'].eq('').sum())
print("NaN:", df['geral'].isna().sum())
```

```python
df.to_csv('data/cbf_df.csv', index=False)
print("Salvo em data/cbf_df.csv")
```
