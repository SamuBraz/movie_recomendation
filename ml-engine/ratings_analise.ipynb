```python
import pandas as pd

rating = pd.read_csv('ml-latest-small/ratings.csv')
movies = pd.read_csv('data/cbf_df.csv')

lista = movies.loc[:, "movieId"].to_list()

###Para essa analise será deconsiderado o timestamp
rating.drop(columns='timestamp')
rating_grouped_by = rating.groupby('movieId')
count_ratings_per_movies = rating_grouped_by.size()

df = count_ratings_per_movies.reset_index(name='count_ratings').sort_values('count_ratings')
df.to_csv('ratings_contandos.csv', index=False)

```

```python
rating_contandos = pd.read_csv('ratings_contandos.csv')
movies = pd.read_csv('data/cbf_df.csv')

movies_ratings = movies.merge(rating_contandos, 'left', 'movieId')

movies_ratings.loc[movies_ratings['count_ratings'].isna()]

movies_ratings['count_ratings'] = movies_ratings['count_ratings'].fillna('0')

movies_ratings = movies_ratings.drop(columns=['genres','tag' ,'year', 'geral'])
movies_ratings['count_ratings'] = movies_ratings['count_ratings'].astype(int)
movies_ratings.to_csv('data/movies_ratings_count.csv')
```
