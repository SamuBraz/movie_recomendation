```python
import pandas as pd
recomendados = pd.read_csv('validation/recomendando.csv')

df_movies = pd.read_csv('data/cbf_df.csv')
df_movies = df_movies.drop(columns=['geral', 'tag'])


df_testing = pd.read_csv('validation/testing.csv')


df_traning = pd.read_csv('validation/training.csv')



recomendados = recomendados.merge(df_movies, 'left', on='movieId').drop(columns=['titulo_movie_lens_y'])
recomendados = recomendados.rename(columns={'titulo_movie_lens_x': 'titulo_movie_lens'})


df_testing = df_testing.merge(df_movies, 'left', on='movieId')
df_traning = df_traning.merge(df_movies, 'left', on='movieId')

# recomendados
# df_testing
# df_traning
```

```python
dummies = recomendados['genres'].str.get_dummies(sep='|')
recomendados = pd.concat([recomendados, dummies], axis=1)
recomendados


dummies = df_testing['genres'].str.get_dummies(sep='|')
df_testing = pd.concat([df_testing, dummies], axis=1)
df_testing

dummies = df_traning['genres'].str.get_dummies(sep='|')
df_traning = pd.concat([df_traning, dummies], axis=1)
df_traning
```

```python
recomendados_generos = recomendados.iloc[:, 4:]
recomendados_generos

df_traning_generos = df_traning.iloc[:, 6:]
df_traning_generos

df_testing_generos = df_testing.iloc[:, 6:]
df_testing_generos


recomendados_generos.sum().sort_values().to_csv('validation/resultado/worst_user/soma_recomendados.csv')
df_traning_generos.sum().sort_values().to_csv('validation/resultado/worst_user/soma_treino.csv')

df_testing_generos.sum().sort_values().to_csv('validation/resultado/worst_user/soma_test.csv')
```

```python
recomendados = recomendados.iloc[:, :4].to_csv('validation/resultado/worst_user/lista/lista_filmes_recomendandos.csv')
df_traning.iloc[:, :6].to_csv('validation/resultado/worst_user/lista/trainng.csv')
df_testing.iloc[:, :6].to_csv('validation/resultado/worst_user/lista/testing.csv')
```

```python
lista_move_id = recomendados.loc[:, "movieId"].to_list()


ratings = pd.read_csv('ml-latest-small/ratings.csv')



ratings_lista = ratings.loc[ratings['movieId'].isin(lista_move_id)]

ratings_lista = ratings_lista.sort_values('movieId')

ratings_lista.to_csv('validation/resultado/worst_user/lista/ratings.csv')
```

```python
groups_ratings = ratings_lista.groupby('movieId')


contagem_avaliacao = groups_ratings.size()
contagem_avaliacao.to_csv('avaliacoes_contagem.csv')
```
