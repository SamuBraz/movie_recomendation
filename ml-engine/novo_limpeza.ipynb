```python
import pandas as pd


relevancia = pd.read_csv('relevancia_filmes_user.csv')
traing_data = pd.read_csv('data_spliting/testing.csv')


traing_data_grouped = traing_data.groupby('userId')

avaliacao_de_dados = traing_data_grouped.size().reset_index(name='Quantidade de avaliações no dataset de test')

total = relevancia.merge(avaliacao_de_dados, 'left', 'userId')

total.to_csv('comparação_filmes_test_relevantes.csv', index=False)
```
